```python
import pandas as pd
import sys 
import numpy as np
import random
import spacy
spacy.load('en_core_web_sm')
from spacy.lang.en import English
parser = English()
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
en_stop = set(nltk.corpus.stopwords.words('english'))
from nltk.corpus import wordnet as wn
from gensim import corpora
import pickle
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt
import time

file_ = '/home/social-sim/Documents/SocialSimCodeTesting/GCN/sgc/preprocess_DARPA_data/raw_data/twitter_cve/attr_74074.csv'
user_attr = pd.read_csv(file_)#,nrows=1000)
user_attr = user_attr.dropna()
print(user_attr.columns)
cols = ['id_h']
s= user_attr.groupby('id_h')["hashtags.keyword: Descending"].apply(lambda x: "{%s}" % ', '.join(x))
#texts = user_attr["user.description_m.keyword: Descending"]
texts = s#user_attr["hashtags.keyword: Descending"]
print(texts.head())

def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma

def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens

def prepare_text_for_lda(text):
	try:
		tokens = tokenize(text)
		tokens = [token for token in tokens if len(token) > 4]
		tokens = [token for token in tokens if token not in en_stop]
		tokens = [get_lemma(token) for token in tokens]
		return tokens
	except:
		print(pd.isnull(text))
		sys.exit()


text_data = []
for line in texts.values.tolist():
	if not pd.isnull(line):
		tokens = prepare_text_for_lda(line)
		if random.random() > .99:
			print(tokens)
			text_data.append(tokens)
	else:
		continue


dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')

import gensim
#NUM_TOPICS = 15
start, limit, step = 2, 40, 3
multi_coherence_values=[]
for n in range(10):
    coherence_values = []
    for NUM_TOPICS in range(start, limit, step):
        ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=20)
        ldamodel.save('model5.gensim')
        topics = ldamodel.print_topics(num_words=4)
        for topic in topics:
            print(topic)

        dictionary = gensim.corpora.Dictionary.load('dictionary.gensim')
        corpus = pickle.load(open('corpus.pkl', 'rb'))
        lda = gensim.models.ldamodel.LdaModel.load('model5.gensim')
        coherence_model_lda = CoherenceModel(model=lda, texts=text_data, dictionary=dictionary, coherence='c_v')
        coherence_lda = coherence_model_lda.get_coherence()
        coherence_values.append(coherence_lda)
    multi_coherence_values.append(coherence_values) 
a = np.array(multi_coherence_values)
mea = np.mean(a,axis=0)
std = np.std(a,axis=0)
x = range(start, limit, step)
plt.fill_between(x, mea - std,mea + std, alpha=0.3,   color='y')
plt.plot(x, mea,  color='b')
plt.xlabel("Number of topics")
plt.ylabel("Coherence score")
#plt.legend(("coherence_values"), loc='best')
plt.grid()
#plt.show()
plt.savefig('plots/coherence_score_%s.pdf'%time.time())
plt.show()
pyLDAvis.save_html(lda_display)
sys.exit()
print('\nCoherence Score: ', coherence_lda)
x = range(start, limit, step)
plt.plot(x, coherence_values,color='grey')
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.grid()
#plt.show()
plt.savefig('coherence_score_%s.pdf'%time.time())
import pyLDAvis.gensim
lda_display = pyLDAvis.gensim.prepare(lda, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
#pyLDAvis.save_html(lda_display)

```
